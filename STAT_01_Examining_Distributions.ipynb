## Types of Variables

**Variables**


---



1.   Quantitative Variables:

      *   *Discrete*:
      A variable whose values are countable is called a discrete
variable. In other words, a discrete variable can assume
only certain values with no intermediate values.

      *   *Continuous*:
      A variable that can assume any numerical value over a
certain interval or intervals is called a continuous variable.




---



2.   Qualitative or Categorical Variables:

      A variable that cannot assume a numerical value but can be classified into two or more nonnumeric categories is called a qualitative or categorical variable. The data collected on such a variable are called qualitative data.

#### Bar Graph

A graph made of bars whose heights represent the
frequencies of respective categories is called a bar graph.

```python


import matplotlib.pyplot as plt

# Create some data
labels = ['A', 'B', 'C', 'D', 'Others']
values = [10, 5, 8, 4, 3]

# Create a bar chart
plt.bar(labels, values)
plt.title('Bar Chart of Medication Types')
plt.xlabel('Medication Type')
plt.ylabel('Number of Patients')
plt.show()
```

A circle divided into portions that represent the relative
frequencies or percentages of a population or a sample
belonging to different categories is called a pie chart.

```python

# Create a pie chart
plt.pie(values, labels=labels, autopct='%1.1f%%')
plt.title('Pie Chart of Medication Types')
plt.show()
```

### Graphical Presentation of Quantitative Data

#### Histogram

A histogram is a graph in which classes are marked on the
horizontal axis and the frequencies, relative frequencies, or
percentages are marked on the vertical axis. The frequencies,
relative frequencies, or percentages are represented by the
heights of the bars. In a histogram, the bars are drawn
adjacent to each other.

**Example:**

Number of patients in a hospital by Age Groups

```python

# Generate a histogram for age groups of patients.
import pandas as pd

# Create a sample data frame
df = pd.DataFrame({
    "AgeGroup": ["0-10", "11-20", "21-30", "31-40", "41-50", "51-60", "61-70", "71-80", "81-90", "91-100"],
    "Count": [10, 15, 20, 25, 37, 26, 25, 21, 30, 5]
})

# Plot the histogram
df.plot.bar(x="AgeGroup", y="Count", title="Histogram of Age Groups")
plt.show()
```

##### Relative Frequency and Percentage Distributions

$$ \text {Relative frequency of that class} = \frac{\text{frequency of that class}}{\text {sum of all frequencies}}$$

$$\% \text{Relative frequency} = \text{Relative frequency of that class} \times 100$$

```python

# Calculate the relative frequency for each age group
df["Relative Frequency"] = df["Count"] / df["Count"].sum()

# Calculate the percentage relative frequency for each age group
df["% Relative Frequency"] = df["Relative Frequency"] * 100

# Print the results
print(df[["AgeGroup", "Relative Frequency", "% Relative Frequency"]])

# Plot the % Relative Frequency histogram
df.plot.bar(x="AgeGroup", y="Relative Frequency", title="Histogram of Relative Frequency of Age Groups")
plt.show()
```

#### Timeplot


- Often we gather data that comes to us in a sequence over a
period of time, this is called a timeplot.

- A time plot of a variable plots each observation against the
time at which it was measured. Put time on the horizontal
scale (X-axis) and the variable we are measuring on the
vertical scale (Y-axis). Connecting the data points by lines
helps emphasize any change over time.

- Many interesting data sets are time series, for example,
data on temperature during a month, number of car
accidents over the winter and summer time, your tuition
fees over the years etc.

```python
import numpy as np

# Generating data for cost of prescription drugs A and B over time for 20 years
years = np.array(range(2000, 2020))
drug_a_cost = 100 + 5 * np.sin(4 * np.pi * years / 5)
drug_b_cost = 150 + 10 * np.sin(2 * np.pi * years / 5)

# Create a DataFrame
df = pd.DataFrame({
    "Year": years,
    "Drug A Cost": drug_a_cost,
    "Drug B Cost": drug_b_cost
})

# Plot the timeplot
df.plot(x="Year", y=["Drug A Cost", "Drug B Cost"], title="Timeplot of Prescription Drug Costs")
plt.show()

```

A **trend** in a time series is a persistent, long-term rise or
fall. A pattern in a time series that repeats itself at known
regular interval of time is called seasonal variation.

## Interpreting distributions

When can see a distribution (graphical representation) of
our data with ( such as stem and leaf plots, histogram etc.)
we look for

- The overall pattern or any deviations.
- Shape, center and spread.
- Outliers: an unusually small or large data value.
- Peaks, known as modes.

### Center

- We look at Hospital A, B and C. We plot the number of patients on Y axis and their age on X axis

```python
### Hosp A
# Create a dataset with 1000 samples, normally distributed with mean 30 and standard deviation 5
data = np.random.normal(loc=15, scale=5, size=1000)

# Create a Pandas dataframe
df = pd.DataFrame(data, columns=['x'])

# Filter the data to only include values between 0 and 100
df = df[(df['x'] >= 0) & (df['x'] <= 100)]

# Plot the histogram
plt.hist(df['x'], bins=100)
plt.xlabel('Age')
plt.xlim(0, 100)
plt.axvline(x=15, linestyle="--", color="red")
plt.ylabel('# of patients')
plt.title('Age Distribution in Hospital A')
plt.show()




### Hosp B
# Create a dataset with 1000 samples, normally distributed with mean 40 and standard deviation 5
data = np.random.normal(loc=45, scale=5, size=1000)

# Create a Pandas dataframe
df = pd.DataFrame(data, columns=['x'])

# Filter the data to only include values between 0 and 100
df = df[(df['x'] >= 0) & (df['x'] <= 100)]

# Plot the histogram
plt.hist(df['x'], bins=100)
plt.xlabel('Age')
plt.xlim(0, 100)
plt.axvline(x=45, linestyle="--", color="red")
plt.ylabel('# of patients')
plt.title('Age Distribution in Hospital B')
plt.show()




### Hospital C
# Create a dataset with 1000 samples, normally distributed with mean 40 and standard deviation 5
data = np.random.normal(loc=70, scale=5, size=1000)

# Create a Pandas dataframe
df = pd.DataFrame(data, columns=['x'])

# Filter the data to only include values between 0 and 100
df = df[(df['x'] >= 0) & (df['x'] <= 100)]

# Plot the histogram
plt.hist(df['x'], bins=100)
plt.xlabel('Age')
plt.xlim(0, 100)
plt.axvline(x=70, linestyle="--", color="red")
plt.ylabel('# of patients')
plt.title('Age Distribution in Hospital C')
plt.show()
```

### Shape

Shape: symmetrical, uniform, skewed, or bimodal

```python
# prompt: give me the code to control the scewness of a distribution

import numpy as np

# Generate a dataset with 1000 samples, normally distributed with mean 0 and standard deviation 1
data = np.random.normal(loc=10, scale=1, size=1000)

# Control the skewness of the distribution by adding a power of the data to itself
skewness = 10  # Change this value to control the skewness
data = data + data*skewness

# Plot the histogram of the skewed data
plt.hist(data)
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('Histogram of Skewed Data (Skewness = {})'.format(skewness))
plt.show()
```
